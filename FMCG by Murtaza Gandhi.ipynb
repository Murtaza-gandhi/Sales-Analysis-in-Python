```python
import pandas as pd
import numpy as np 
import datetime as dt 
```

```python
df = pd.DataFrame(pd.read_csv("Raw Sales Data.csv"))
```

```python
df.info()
```

```python
pd.DataFrame(df.groupby("Category")["Qty"].sum())
```

```python
df
```

```python
df["Qty"] = df["Qty"].str.replace(",","").astype(int)
df["Sales"] = df["Sales"].str.replace(",","").astype(int)
df['Month'] = pd.to_datetime(df["Month"])
```

# Requariment 1

# Monthwise number of quantities sold

```python
df
```

```python
df['Month-Year']= df['Month'].dt.strftime("%b-%y")
```

```python
monthqty = pd.DataFrame(df.groupby("Month-Year")['Qty'].sum()).reset_index()
monthqty
```

# Requirement 2

# Category wise Qty sold

```python
Category = pd.DataFrame(df.groupby("Category")["Qty"].sum()).reset_index()
Category
```

# Requirement 3

# Sub-channel wise total sales

```python
Sub_channel = pd.DataFrame(df.groupby("Sub-Channels")["Sales"].sum()).reset_index()
Sub_channel
```

# Requirement 4

# Highest sale in which month

```python
monthly_sales = pd.DataFrame(df.groupby('Month-Year')["Sales"].sum()).reset_index()

monthly_sales.sort_values("Sales",ascending=False).head(1)
```

# Requirement 5

# Highest selling product (based on sales)

```python
prdinfo = df.copy()
prdinfo["Product Name"]= prdinfo["Product Name"].str.replace('\d+', '---').str.split("---").str[0]


product_sales = pd.DataFrame(prdinfo.groupby("Product Name")["Sales"].sum()).reset_index()
product_sales.sort_values("Sales",ascending=False).head(1)

```

# Requirement 6

# List of different products company is offering

```python
prd_comp = prdinfo.drop("Sales",axis=1)
prd_comp.sort_values("Product Name",ascending=False)
```
